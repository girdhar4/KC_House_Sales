Creating a DataFrame for storing Evaluation Metrics

```python
evaluation = pd.DataFrame({'Model': [],
                           'Details':[],
                           'RMSE':[],
                           'R-sq (train)':[],
                           'Adjusted R-sq (train)':[],
                           'R-sq (test)':[],
                           'Adjusted R-sq (test)':[]})
```

```python
# Function to calculate adjusted R square
def adjustedR2(r2,n,k):
    return r2-(k-1)/(n-k)*(1-r2)
```

LINEAR REGRESSION

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
living_space = df['sqft_living']
price = df['price']
x = living_space.values.reshape(-1,1)
y = price.values.reshape(-1,1)
X_train, X_test, y_train, y_test = train_test_split(x, y, train_size = 0.8)
```

```python
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

```python
pred = regressor.predict(X_test)
rmsesm = float(format(np.sqrt(metrics.mean_squared_error(y_test,pred)),'.3f'))
rtrsm = float(format(regressor.score(X_train, y_train),'.3f'))
rtesm = float(format(regressor.score(X_test, y_test),'.3f'))
```

```python
r = evaluation.shape[0]
evaluation.loc[r] = ['Simple Linear Regression','-',rmsesm,rtrsm,'-',rtesm,'-']
evaluation
```

```python
import matplotlib.pyplot as plt

plt.scatter(X_train,y_train)
plt.plot(X_train,regressor.predict(X_train), color='red')
plt.title('Training Graph for Housing')
plt.xlabel('Area')
plt.ylabel('Price')
```

```python
#Create a linear Regression Using statsmodels.api
import statsmodels.formula.api as smf
results = smf.ols('price ~ sqft_living', data=df).fit()
print(results.summary())
```

## MULTIPLE REGRESSION -1

Selecting Features - bedrooms, bathrooms, sqft_living, sqft_lot, floors

```python
features1=['bedrooms','bathrooms','sqft_living','sqft_lot','floors']
```

```python
train_data_ml1,test_data_ml1 = train_test_split(df,train_size = 0.8,random_state=3)
ml1 = LinearRegression()
ml1.fit(train_data_ml1[features1],train_data_ml1['price'])
```

```python
pred1 = ml1.predict(test_data_ml1[features1])
rmsecm1 = float(format(np.sqrt(metrics.mean_squared_error(test_data_ml1['price'],pred1)),'.3f'))
rtrcm1 = float(format(ml1.score(train_data_ml1[features1],train_data_ml1['price']),'.3f'))
artrcm1 = float(format(adjustedR2(ml1.score(train_data_ml1[features1],train_data_ml1['price']),train_data_ml1.shape[0],len(features1)),'.3f'))
rtecm1 = float(format(ml1.score(test_data_ml1[features1],test_data_ml1['price']),'.3f'))
artecm1 = float(format(adjustedR2(ml1.score(test_data_ml1[features1],test_data_ml1['price']),test_data_ml1.shape[0],len(features1)),'.3f'))
```

```python
r = evaluation.shape[0]
evaluation.loc[r] = ['Multiple Regression-1','selected features',rmsecm1,rtrcm1,artrcm1,rtecm1,artecm1]
evaluation.head()
```

```python
#Create a Multiple Regression Using statsmodels.api and displaying the summary
import statsmodels.api as sm
model = sm.OLS(train_data_ml1['price'], sm.add_constant(train_data_ml1[features1])).fit()
Y_pred = model.predict(sm.add_constant(test_data_ml1[features1]))
print_model = model.summary()
print(print_model)
```

## Decision Tree

```python
from sklearn.preprocessing import MinMaxScaler
# define min max scaler
scaler = MinMaxScaler()
# transform data
scaled = scaler.fit_transform(df)
scaled_df= pd.DataFrame(scaled)
scaled_df.head()
```

```python
X = scaled_df.iloc[:, 1:18]
y = scaled_df.iloc[:,0]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=33)
```

```python
from sklearn.tree import DecisionTreeRegressor
tr_regressor = DecisionTreeRegressor(random_state=0)
tr_regressor.fit(X_train,y_train)
tr_regressor.score(X_test,y_test)
```

```python
print("Decision tree  Regression Model Score is ",round(tr_regressor.score(X_test,y_test)*100))
```

```python
pred_dt = tr_regressor.predict(X_test)
rmsecm5 = float(format(np.sqrt(metrics.mean_squared_error(y_test,pred_dt)),'.3f'))
rtrcm5 = float(format(tr_regressor.score(X_train,y_train),'.3f'))
artrcm5 = float(format(adjustedR2(tr_regressor.score(X_train,y_train),X_train.shape[0],X_train.shape[1]),'.3f'))
rtecm5 = float(format(tr_regressor.score(X_test,y_test),'.3f'))
artecm5= float(format(adjustedR2(tr_regressor.score(X_test,y_test),X_train.shape[0],X_train.shape[1]),'.3f'))
```

```python
r = evaluation.shape[0]
evaluation.loc[r] = ['Decision Tree','All features',rmsecm5,rtrcm5,artrcm5,rtecm5,artecm5]
evaluation
```
